# MODELE EKONOMETRYCZNE

## wszystkie

```python
df_ekon_w = pd.read_csv('datasets/cale/ekonometryczne_wszystkie_2019_03_28__12_17_36.csv', index_col=0)
df_ekon_w = df_ekon_w.astype('float64')
df_ekon_w.index = pd.to_datetime(df_ekon_w.index)
print(df_ekon_w.index)
print(df_ekon_w.dtypes)
```

```python
train_ekon_w = df_ekon_w[df_ekon_w.index >= '2012-07-02']
train_ekon_w = train_ekon_w[train_ekon_w.index < '2018-01-01']
valid1_ekon_w = df_ekon_w[df_ekon_w.index >= '2018-01-01']
valid1_ekon_w = valid1_ekon_w[valid1_ekon_w.index < '2018-03-01']
valid2_ekon_w = df_ekon_w[df_ekon_w.index >= '2018-03-01']
valid2_ekon_w = valid2_ekon_w[valid2_ekon_w.index < '2018-05-01']
valid3_ekon_w = df_ekon_w[df_ekon_w.index >= '2018-05-01']
valid3_ekon_w = valid3_ekon_w[valid3_ekon_w.index < '2018-07-01']
test_ekon_w = df_ekon_w[df_ekon_w.index >= '2018-07-01']
test_ekon_w = test_ekon_w[test_ekon_w.index < '2019-01-01']
```

```python
print('----------------------------------------------- TRAIN ---------------------------------------------------------------')
display(train_ekon_w.head(3))
display(train_ekon_w.tail(3))
print(train_ekon_w.shape)
print('----------------------------------------------- VALID1 ---------------------------------------------------------------')
display(valid1_ekon_w.head(3))
display(valid1_ekon_w.tail(3))
print(valid1_ekon_w.shape)
print('----------------------------------------------- VALID2---------------------------------------------------------------')
display(valid2_ekon_w.head(3))
display(valid2_ekon_w.tail(3))
print(valid2_ekon_w.shape)
print('----------------------------------------------- VALID3 ---------------------------------------------------------------')
display(valid3_ekon_w.head(3))
display(valid3_ekon_w.tail(3))
print(valid3_ekon_w.shape)
print('------------------------------------------------ TEST --------------------------------------------------------------')
display(test_ekon_w.head(3))
display(test_ekon_w.tail(3))
print(test_ekon_w.shape)
```

```python
train_ekon_w.to_csv('datasets/podzielone/ekonometryczne/wszystkie/train_ekon.csv')
valid1_ekon_w.to_csv('datasets/podzielone/ekonometryczne/wszystkie/valid1_ekon.csv')
valid2_ekon_w.to_csv('datasets/podzielone/ekonometryczne/wszystkie/valid2_ekon.csv')
valid3_ekon_w.to_csv('datasets/podzielone/ekonometryczne/wszystkie/valid3_ekon.csv')
test_ekon_w.to_csv('datasets/podzielone/ekonometryczne/wszystkie/test_ekon.csv')
```
