```python
import pandas as pd
```

```python
ekon_wszystkie = pd.read_csv('datasets/cale/ekonometryczne_wszystkie.csv', index_col=0)
ekon_wszystkie_upd = pd.read_csv('datasets/cale/ekonometryczne_wszystkie_2019_03_28__11_52_34.csv', index_col=0)
df_lag = pd.read_csv('datasets/cale/zbior_do_lagow.csv', index_col=0)
ml_wszystkie = pd.read_csv('datasets/cale/ml_wszystkie.csv', index_col=0)

display(ekon_wszystkie.head())
display(ekon_wszystkie_upd.head())
display(df_lag.head())
display(ml_wszystkie.head())
```

```python
not_found = []
for col in ekon_wszystkie.columns.drop('return ratio'):
    try:
        ekon_wszystkie[col] = df_lag[col].shift(1).loc[ekon_wszystkie.index]
    except Exception as e:
        not_found.append(col)
        
for col in not_found:
    try:
        ekon_wszystkie[col] = ml_wszystkie[str(col)]
    except Exception as e:
        print(col)
```

```python
ekon_wszystkie.to_csv('datasets/cale/ekonometryczne_wszystkie_2019_03_28__12_17_36.csv')
```

```python
ekon_st = pd.read_csv('datasets/cale/ekonometryczne_stacjonarne.csv', index_col=0)
```
